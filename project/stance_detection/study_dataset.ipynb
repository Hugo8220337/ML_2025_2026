```python
import sys
from pathlib import Path
import pandas as pd

# Add project directory to Python path
project_dir = Path().resolve().parent
sys.path.insert(0, str(project_dir))

from common.tools import read_csv

BODY_DATA_PATH = project_dir / "datasets" / "fnc1" / "fnc1_train_bodies.csv"
STANCES_DATA_PATH = project_dir / "datasets" / "fnc1" / "fnc1_train_stances.csv"
```

```python
bodies = read_csv(BODY_DATA_PATH)
stances = read_csv(STANCES_DATA_PATH)

# Merge bodies and stances on 'Body ID'
df = pd.merge(stances, bodies, on='Body ID')
```

```python
# Map tables to numbers
label_mapping = {
    'agree': 0,
    'disagree': 1,
    'discuss': 2,
    'unrelated': 3
}
df['label'] = df['Stance'].map(label_mapping)

# saber quantidade de labels por label num gr√°fico
label_counts = df['label'].value_counts().sort_index()
print("Label Counts:")
print(label_counts)
```
